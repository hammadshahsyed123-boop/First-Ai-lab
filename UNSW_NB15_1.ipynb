```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;

file_path = '/content/drive/MyDrive/AI Lab/UNSW-NB15_1.csv'
print(file_path)

```

```python
data = pd.read_csv(file_path)
print(data.head())

```

```python
# Assuming 'udp' is the column name, replace it if it's different based on your data's actual column names
# You can verify the column names using data.columns

# Bar chart for 'udp' column
plt.figure(figsize=(10, 6))
data['udp'].value_counts().plot(kind='bar')
plt.title('Distribution of UDP values')
plt.xlabel('UDP values')
plt.ylabel('Count')
plt.show()

# Pie chart for 'udp' column
plt.figure(figsize=(8, 8))
data['udp'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of UDP values')
plt.ylabel('') # Remove default y-label for pie chart
plt.show()
```
