# **Task 5: Numpy**

Reproduced Result of **Task 5**

```python
import numpy as np

# Create a different 1D array
arr_new = np.arange(100, 120)
print('New 1D Array:', arr_new)

# Reshape to 4x5 matrix
matrix_new = arr_new.reshape(4, 5)
print('New 4x5 Matrix:\n', matrix_new)

# Mean, Median, Std
print('Mean:', np.mean(arr_new))
print('Median:', np.median(arr_new))
print('Standard Deviation:', np.std(arr_new))

# Element-wise operations with a different array
arr_new_2 = np.arange(120, 140)
print('Addition:', arr_new + arr_new_2)
print('Multiplication:', arr_new * arr_new_2)
```

# **Task 6: Matplotlib**

```python
import matplotlib.pyplot as plt

# Line Plot
x_new = list(range(5, 15))
y_new = [i*3 for i in x_new]  # Using a different function for y
plt.plot(x_new, y_new)
plt.title('Numbers vs Triples')
plt.xlabel('Numbers')
plt.ylabel('Triples')
plt.show()

# Bar Chart
students_new = ['David', 'Emily', 'Michael', 'Sophia', 'Daniel']
marks_new = [75, 88, 95, 80, 92] # Different marks
plt.bar(students_new, marks_new)
plt.title('New Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()

# Pie Chart
categories_new = ['Category X', 'Category Y', 'Category Z']
values_new = [40, 35, 25] # Different distribution
plt.pie(values_new, labels=categories_new, autopct='%1.1f%%')
plt.title('New Category Distribution')
plt.show()
```
